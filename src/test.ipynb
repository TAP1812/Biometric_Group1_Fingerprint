```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../dataset/DB2_B/101_1.tif', 0)
```

```python


# HOG Calculation
def calculate_hog(image):
    winSize = (64,64)
    blockSize = (16,16)
    blockStride = (8,8)
    cellSize = (8,8)
    nbins = 9

    hog = cv.HOGDescriptor(winSize, blockSize, blockStride, cellSize, nbins)
    hist = hog.compute(image)
    hist_flat = hist.flatten()  # Flattening the HOG features into a vector
    return hist_flat

```

```python
W = 16 
hog_hist = calculate_hog(img) 
print(f'HOG Features: {len(hog_hist)}')
```
